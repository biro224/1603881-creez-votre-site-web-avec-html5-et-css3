```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
trains=pd.read_csv('real_estate.csv')
print(trains)
```

```python
print(trains.head())
```

```python
plt.plot(trains['No'],trains['Age'])
```

```python
plt.plot(trains['Transact Date'],trains['Price'])
plt.xlabel('Transact Date')
plt.ylabel('Price')
plt.title('Price over Time')

plt.show()
```

```python
sns.lineplot(x='Transact Date', y='Price',data=trains)
plt.xlabel('Transact Date')
plt.ylabel('Price')
plt.title('Price over Time')
plt.show()
```

```python
plt.bar(trains['Transact Date'],trains['Price'])
plt.xlabel('Transact Date')
plt.ylabel('Price')
plt.show()
```

```python
sns.countplot(x='Transact Date',data=trains)
plt.xlabel('Transact Date')
plt.ylabel('Price')
```
